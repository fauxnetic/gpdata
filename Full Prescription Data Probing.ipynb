```python
data_file = './full_data/2010_08_August/T201008PDPI BNFT.CSV'
data_col_names = ['SHA', 'PCT', 'PracticeCode', 'BNFCode', 'BNFName',
                  'LineCountDispensed', 'NetCost','ActualCost','QuantityDispensed', 'Year-Month']
data = pd.read_csv(data_file, names = data_col_names, skiprows = 1, index_col=False)
data['Year-Month'] = pd.to_datetime(data['Year-Month'], format='%Y%m')
data
```

# Difference in Net and Actual costs

```python
data['ActualCostDiff'] = data['NetCost'] - data['ActualCost']
# Total Cost to practice
data['TotalCost'] = data['ActualCost'] * data['LineCountDispensed']
data
```

```python
plt.boxplot(data['ActualCostDiff'])
plt.yscale('log')
plt.show()
```

## Medicinces with largest changes in cost to prescribe
- By raw numbers
- By percentage cost

```python
maxCostDiff = data['ActualCostDiff'].max()
data[data['ActualCostDiff'] == maxCostDiff]
```

```python
minCostDiff = data['ActualCostDiff'].min()
data[data['ActualCostDiff'] == minCostDiff]
```

```python
data['PercentSaving'] = data['NetCost'] / data['ActualCostDiff']
```

```python
plt.violinplot(data['ActualCostDiff'], showmeans=False, showmedians=True)
plt.yscale('log')
plt.show()
```
