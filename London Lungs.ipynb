## Get London & non-London Practices

```python
addresses = px.load_address_data(2015, 10)
list(addresses.columns.values)
```

```python
londonPractices = addresses[addresses['PostCode'].str.contains('^(N|NW|W|SW|SE|E|WC|EC)[0-9]')]
londonPractices
```

# Partition Data

```python
all_px_from_london = pxdata[pxdata['PracticeCode'].isin(londonPractices['PracticeCode'])]
all_px_from_elsewhere = pxdata[~pxdata['PracticeCode'].isin(londonPractices['PracticeCode'])]

inhalers_from_london = all_px_from_london[all_px_from_london['BNFCode'] == '0301011R0AAAPAP']
inhalers_from_elsewhere = all_px_from_elsewhere[all_px_from_elsewhere['BNFCode'] == '0301011R0AAAPAP']
```

```python
len(all_px_from_london.index)
```

```python
len(all_px_from_elsewhere)
```

```python
len(inhalers_from_london)
```

```python
len(inhalers_from_elsewhere)
```

```python
total_px_london = all_px_from_london['LineCountDispensed'].sum()
total_px_elsewhere = all_px_from_elsewhere['LineCountDispensed'].sum()

total_inhalers_london = inhalers_from_london['LineCountDispensed'].sum()
total_inhalers_elsewhere = inhalers_from_elsewhere['LineCountDispensed'].sum()
```

```python
total_px_london
```

```python
total_px_elsewhere
```

```python
total_inhalers_london
```

```python
total_inhalers_elsewhere
```

```python
percentage_london = (total_inhalers_london / total_px_london) * 100
percentage_london
```

```python
percentage_elsewhere = (total_inhalers_elsewhere / total_px_elsewhere) * 100
percentage_elsewhere
```

```python
from scipy import stats
import numpy as np
```

```python
inhaler_proportions = np.array([[total_inhalers_london, total_px_london], [total_inhalers_elsewhere, total_px_elsewhere]])
```

```python
[chi2, pval, dof, expected] = stats.chi2_contingency(inhaler_proportions)
```
